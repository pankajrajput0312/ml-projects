```python
def dist(x1,x2):
    return np.sqrt(sum(x1-x2)**2)
```

```python
def KNN(x,y,query,k=5):
    vals=[]
    m=x.shape[0]
    for i in range(m):
        d=dist(query,x[i])
        vals.append((d,y[i]))
    vals=sorted(vals)
    vals=vals[:k]
    vals=np.array(vals)
    newvals=np.unique(vals[:,1],return_counts=True)
    max_freq_index=newvals[1].argmax()
    pred=newvals[0][max_freq_index]
    return pred
```
